```
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
def f1(x):
    return -np.sqrt(9 -(-6-x)**2)
x = np.linspace(-9,-6,100)
y = f1(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f2(x):
   return 1*x+3
x = np.linspace(-6,-3,100)
y = f2(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f3(x):
    return np.sqrt(9-(x-0)**2)
x = np.linspace(-3,0,100)
y = f3(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f4(x):
    return -1*x +3
x = np.linspace(0,3,100)
y = f4(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f5(x):
   return 0.5*x-1.5
x = np.linspace(3,9,100)
y = f5(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def foo(x):
   if (x >= -9 and x < -6 ):
        return f1(x)
   elif (x >= -6 and x < -3 ):
        return f2(x)
   elif (x >= -3 and x < 0 ):
        return f3(x)
   elif (x >= 0 and x < 3 ):
        return f4(x)
   elif (x >= 3):
        return f5(x)
x = np.linspace(-9,8,1000)
y = [foo(i) for i in x]
plt.figure(figsize=(12,4))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```
